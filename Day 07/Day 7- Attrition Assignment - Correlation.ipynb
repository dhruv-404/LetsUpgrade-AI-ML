```python
import pandas as pd

dataset1=pd.read_csv('general_data.csv')
dataset1.head()
```

```python
from scipy.stats import pearsonr
```

```python
dataset1.drop_duplicates()
```

```python
##correlation between distance from home and education
stats,p=pearsonr(dataset1.DistanceFromHome,dataset1.Education)
print(stats,p)
```

```python
##correlation between monthly income and eduacation
stats,p=pearsonr(dataset1.MonthlyIncome,dataset1.Education)
print(stats,p)
```

```python
##correlation between attrition and education
stats,p=pearsonr(dataset1.Attrition,dataset1.Education)
print(stats,p)
```
